```python
import nltk
'''
5 things to do with NLTK -
a. tokenization
b. stopword removal
c. stemming
d. lemmatization
'''
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
raw_text = "I am learning python programming ,which I would be using in generative ai and creating realtime agents... !!"
# Step 1 :- Lowercase
raw_text = raw_text.lower()
print("After lower case : ",raw_text)

# Directly Spliting karte wqt coma vghera word ke sath aa jate hai :-
print("After spliting : ", raw_text.split())

# Step 2 :-Tokenization
tokens = word_tokenize(raw_text)
print("After tokenization : ",tokens)
```

```python
# Step 3:- Remove stopwords
eng_stopwords = stopwords.words("english")
print("English stopwords : ",eng_stopwords)
```

```python
filtered_token = []
for word in tokens:
  if word not in eng_stopwords:
    filtered_token.append(word)
print("After removing stopwords : ",filtered_token)
```

```python
# Step 4 :- Remove Punctuation
import string
puncuations = string.punctuation
print("Puncuations : ",puncuations)

clean_token = [word for word in filtered_token if word not in puncuations]
print("After removing puncuations : ",clean_token)
```

```python
# Step 5 = Stemmming :- remove -ing
stem = PorterStemmer()
stem.stem("creating")
```

```python
#step 6 :- lemmatizing
wnet = WordNetLemmatizer()
wnet.lemmatize("creating","v")
wnet.lemmatize("palying","v")
wnet.lemmatize("running","v")
wnet.lemmatize("went","v")
```

```python
lemmatized_word =[]
for word in clean_token:
  lemmatized_word.append(wnet.lemmatize(word,"v"))
print("Lemmatized Word: ",lemmatized_word)
```

```python
final_word = []
for word in lemmatized_word:
  if word.isalpha():
    final_word.append(word)
print("Final Word: ",final_word)
```

```python
'''
TASK 1 -
'''
sentences = [
    "I am learning python programming ",
    "We are happy Today",
    "I am sad today"
]
def preprocess_text(sentences):
  eng_stopwords = stopwords.words("english")
  puncuations = string.punctuation
  all_token = []
  for sentence in sentences:
      sentence = sentence.lower()
      token = word_tokenize(sentence)
      filtered_tokens = []
      for word in token:
        if word not in eng_stopwords:
          filtered_tokens.append(word)
      print("After filtering : ",filtered_tokens)

      clean_token = [word for word in filtered_tokens if word not in puncuations]
      lemmatized_word =[]
      for word in clean_token:
        lemmatized_word.append(wnet.lemmatize(word,"v"))
      final_words = []
      for word in clean_token:
        if word.isalpha():
          final_words.append(word)
      all_token.append(final_words)
  return all_token

cleaned_sentence = preprocess_text(sentences)
print("Cleaned Sentence : ",cleaned_sentence)
```
